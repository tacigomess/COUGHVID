```python
import os
import matplotlib.pyplot as plt
import librosa
import librosa.display
import IPython.display as ipd

#audio_fpath = "/content/content/audios_segmentados_concatenados_especialistas"
#audio_fpath = "/content/content/audios_segmentados_concatenados_especialistas_plus_adicional"
#audio_fpath = "/content/audios_train_especialistas_plus_adicional"
#audio_fpath = "/content/content/audios_trim40_train_especialistas"
#audio_fpath = "/content/content/audios_trim90_especialistas"
#audio_fpath = "/content/content/audios_trim60_especialistas"
#audio_fpath = "/content/content/audios_trim40_especialistas"
#audio_fpath = "/content/content/audios_expecialistas"
#audio_fpath = "/content/content/audios_especialistas_plus_adicional_8750"
#audio_fpath = "/content/content/audios_segmentados_concatenados_especialistas_completo"
#audio_fpath = "/content/content/audios_segmentados_concatenados_especialistas_completo_default_parameters"
audio_fpath = "/content/content/audios_segmentados_concatenados_especialistas_plus_adicional_completo"


audio_clips = os.listdir(audio_fpath)
print("No. of .wav files in audio folder = ",len(audio_clips))
```

```python
import soundfile as sf
import numpy as np
import pylab

!mkdir audios_MFCC_segmentados_concatenados_especialistas_plus_adicional_completo
audio_fpath_destinoMFCC = "/content/audios_MFCC_segmentados_concatenados_especialistas_plus_adicional_completo"

!mkdir audios_MEL_segmentados_concatenados_especialistas_plus_adicional_completo
audio_fpath_destinoMEL = "/content/audios_MEL_segmentados_concatenados_especialistas_plus_adicional_completo"

for i in audio_clips:
      x, fs = librosa.load(audio_fpath+'/'+i, sr=None)  

      # extract mfcc
      mfcc = librosa.feature.mfcc(x, fs, n_mfcc=13, n_fft=2048, hop_length=512)
      mfcc = mfcc.T
      imageMFCC = librosa.power_to_db(mfcc, ref=np.max)

      pylab.figure(figsize=(3,3))
      pylab.axis('off') 
      pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[]) # Remove the white edge
      librosa.display.specshow(imageMFCC)
      i = i.replace("wav","jpg")
      pylab.savefig(audio_fpath_destinoMFCC +'/'+ i, bbox_inches=None, pad_inches=0)
      pylab.close()

      # extract mel spectogram
      mel_spect = librosa.feature.melspectrogram(y=x, sr=fs, n_fft=2048, hop_length=1024)
      mel_spect = librosa.power_to_db(mel_spect, ref=np.max)

      pylab.figure(figsize=(3,3))
      pylab.axis('off') 
      pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[]) # Remove the white edge
      librosa.display.specshow(mel_spect)
      pylab.savefig(audio_fpath_destinoMEL +'/'+ i, bbox_inches=None, pad_inches=0)
      pylab.close()
```
