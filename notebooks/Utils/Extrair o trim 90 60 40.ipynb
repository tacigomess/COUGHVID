```python
import os
import matplotlib.pyplot as plt
import librosa
import librosa.display
import IPython.display as ipd

#audio_fpath = "/content/content/audios_segmentados_concatenados_especialistas"
#audio_fpath = "/content/content/audios_segmentados_concatenados_especialistas_plus_adicional"
audio_fpath = "/content/content/audios_expecialistas"

audio_clips = os.listdir(audio_fpath)
print("No. of .wav files in audio folder = ",len(audio_clips))
```

```python
import soundfile as sf
import numpy as np
import pylab

!mkdir audios_trim90_especialistas
audio_fpath_destino_trim90 = "/content/audios_trim90_especialistas"

!mkdir audios_trim60_especialistas
audio_fpath_destino_trim60 = "/content/audios_trim60_especialistas"

!mkdir audios_trim40_especialistas
audio_fpath_destino_trim40 = "/content/audios_trim40_especialistas"

for i in audio_clips:
      x, fs = librosa.load(audio_fpath+'/'+i, sr=None)  
      trim90 = librosa.effects.trim(x, top_db=90)[0]
      trim60 = librosa.effects.trim(x, top_db=60)[0]
      trim40 = librosa.effects.trim(x, top_db=40)[0]

      sf.write(audio_fpath_destino_trim90+"/" +i, data= trim90, samplerate=fs, subtype='PCM_24')
      sf.write(audio_fpath_destino_trim60+"/" +i, data= trim60, samplerate=fs, subtype='PCM_24')
      sf.write(audio_fpath_destino_trim40+"/" +i, data= trim40, samplerate=fs, subtype='PCM_24')
```
