```python
#Filtro para pegar s os audios que estiverem descritos no df_train_original
import soundfile as sf
import glob
import os
from os import listdir
from os.path import isfile, join
import audioread

def filtroAudiosWAV(audio_fpath,audio_clips,df,destino):

  for y in range(0,len(df)):
    try: 
        
      destino_local = destino+df['uuid'].iloc[y]
      origem_local = audio_fpath+df['uuid'].iloc[y]

      y, sr = librosa.load(origem_local+'.wav')
      sf.write(destino_local+'.wav', data= y, samplerate=sr, subtype='PCM_24')

    except RuntimeError:
      print("arquivo escondido")
      print("RuntimeError")
    except:
      print("Something else went wrong")
```
