```python
#Filtro para pegar só os audios que estiverem descritos no df_expert
import soundfile as sf
import glob
import os
from os import listdir
from os.path import isfile, join
import audioread

audio_fpath = "/content/audio_wav/"
audio_clips = os.listdir(audio_fpath)
!mkdir audios_especialistas
def filtroAudiosEspecialistas(audio_fpath,audio_clips,df_expert):

  audio_fpath_destinoEspecialista = "/content/audios_especialistas/"

  for y in range(0,len(df_expert)):
    try: 
        
      destino_local = audio_fpath_destinoEspecialista+df_expert['uuid'].iloc[y]
      origem_local = audio_fpath+df_expert['uuid'].iloc[y]

      y, sr = librosa.load(origem_local+'.wav')
      sf.write(destino_local+'.wav', data= y, samplerate=sr, subtype='PCM_24')

    except RuntimeError:
      print("arquivo escondido")
      print("RuntimeError")
    except:
      print("Something else went wrong")
```

```python
filtroAudiosEspecialistas(audio_fpath,audio_clips,df_expert)
```
